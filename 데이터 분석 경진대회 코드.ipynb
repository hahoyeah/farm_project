## 나이

### 전체 나이 비율

```python
age=hurt.loc[:,['id','q10']]
age.head()
```

```python
age2 = age.groupby(['q10'], as_index=False).count()
age2 = age2.sort_values(by=['q10'])
age2.head()
```

```python
age2.loc[age2['q10'] <= 39, '범주'] = "40세 미만"
age2.loc[(age2['q10'] >= 40) & (age2['q10'] <= 49), '범주'] = "40대"
age2.loc[(age2['q10'] >= 50) & (age2['q10'] <= 59), '범주'] = "50대"
age2.loc[(age2['q10'] >= 60) & (age2['q10'] <= 69), '범주'] = "60대"
age2.loc[age2['q10'] >= 70, '범주'] = "70세 이상"
age2.head()
```

```python
ratio = age2["범주"].value_counts()
ratio
```

```python
fig = plt.figure(figsize=(10,10))
plt.pie(ratio, labels=['70세 이상', '40세 미만', '60대', '40대', '50대'], autopct='%.1f%%', startangle=260, counterclock=False, textprops={'fontsize':20})
plt.title("전체 나이 비율", fontsize=40)
plt.show()
```

## 농작업시 위험 정도와 농작업 시 안전 신경 정도

### 손상자

```python
danger=hurt_plus2.loc[:,['q20','q22']]
danger.head()
```

```python
danger20=danger.groupby(['q20'],as_index=False).count()
danger20=danger20.rename(columns={'q22':'count'})
danger20
```

```python
danger22=danger.groupby(['q22'],as_index=False).count()
danger22=danger22.rename(columns={'q20':'count'})
danger22
```
